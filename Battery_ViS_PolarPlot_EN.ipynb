```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import matplotlib as mpl

plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(18, 12))

# mpl.rcParams['font.sans-serif']=['SimHei']
# mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.size'] = 14.0

font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc', size = 12)

ability_size = 6
ability_label = ['Energy Density', 'Power Density', 'Safety', 'Cycle Life', 'Price/Ah','Performance']


ax1 = plt.subplot(231, projection = 'polar')
ax2 = plt.subplot(232, projection = 'polar')
ax3 = plt.subplot(233, projection = 'polar')
ax4 = plt.subplot(234, projection = 'polar')
ax5 = plt.subplot(235, projection = 'polar')
ax6 = plt.subplot(236, projection = 'polar')

LCO = [4, 2, 2, 2, 1, 3]
NCA = [4, 3, 2, 3, 2, 3]
NMC = [4, 3, 3, 3, 3, 3]
LMO = [3, 3, 3, 2, 3, 2]
LFP = [2, 4, 4, 4, 3, 3]
LTO = [2, 3, 4, 4, 2, 4]



theta = np.linspace(0, 2*np.pi, 6, endpoint = False)

theta = np.append(theta,theta[0])

LCO = np.append(LCO, LCO[0])
ax1.plot(theta, LCO, 'r')
ax1.fill(theta, LCO, 'r', alpha = 0.3)
ax1.set_xticks(theta)
ax1.set_xticklabels(ability_label, fontproperties=font, y=0.005, size = 16)
ax1.set_title('LCO', position=(0.5, 1.04), fontproperties=font, color = 'r',size = 20)
ax1.set_yticks([1, 2, 3, 4, 5])

NCA = np.append(NCA, NCA[0])
ax2.plot(theta, NCA, 'orange')
ax2.fill(theta, NCA, 'orange', alpha = 0.3)
ax2.set_xticks(theta)
ax2.set_xticklabels(ability_label, fontproperties=font, y=0.005, size = 16)
ax2.set_title('NCA', position=(0.5, 1.04), fontproperties=font, color = 'r',size = 20)
ax2.set_yticks([1, 2, 3, 4, 5])

NMC = np.append(NMC, NMC[0])
ax3.plot(theta, NMC, 'b')
ax3.fill(theta, NMC, 'b', alpha = 0.3)
ax3.set_xticks(theta)
ax3.set_xticklabels(ability_label, fontproperties=font, y=0.005, size = 16)
ax3.set_title('NMC', position=(0.5, 1.04), fontproperties=font, color = 'r',size = 20)
ax3.set_yticks([1, 2, 3, 4, 5])

LMO = np.append(LMO, LMO[0])
ax4.plot(theta, LMO, 'y')
ax4.fill(theta, LMO, 'y', alpha = 0.3)
ax4.set_xticks(theta)
ax4.set_xticklabels(ability_label, fontproperties=font, y=0.005, size = 16)
ax4.set_title('LMO', position=(0.5, 1.04), fontproperties=font, color = 'r',size = 20)
ax4.set_yticks([1, 2, 3, 4, 5])

LFP = np.append(LFP, LFP[0])
ax5.plot(theta, LFP, 'g')
ax5.fill(theta, LFP, 'g', alpha = 0.3)
ax5.set_xticks(theta)
ax5.set_xticklabels(ability_label, fontproperties=font, y=0.005, size = 16)
ax5.set_title('LFP', position=(0.5, 1.04), fontproperties=font, color = 'r',size = 20)
ax5.set_yticks([1, 2, 3, 4, 5])

LTO = np.append(LTO , LTO [0])
ax6.plot(theta, LTO , 'purple')
ax6.fill(theta, LTO , 'purple', alpha = 0.3)
ax6.set_xticks(theta)
ax6.set_xticklabels(ability_label, fontproperties=font, y=0.005, size = 16)
ax6.set_title('LTO', position=(0.5, 1.04), fontproperties=font, color = 'r',size = 20)
ax6.set_yticks([1, 2, 3, 4, 5])

fig.suptitle('Commercial Cathode Electrodes Comparison', fontsize=22, y =1.005)
plt.tight_layout()
plt.show()
```
